```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
import statsmodels.api as sm
from patsy import dmatrices
%matplotlib inline
```

```python
df = pd.read_csv("Auto.csv")
df.columns
```

```python
df.columns
```

1. Year as numerical variable
--------------

```python
y, X = dmatrices('mpg ~ year', data=df, return_type='dataframe')
model = sm.OLS(y,X)
result = model.fit()
```

```python
ax = df.plot(x='year', y='mpg', kind='scatter')

#Wykres bez sensu
p = result.params
x = np.arange(65,85,1)
ax = ax.plot(x, p.Intercept + p.year * x, 'r')
```

2. Treating year as a categorical variable.
------------

```python
df['mpg'] = (df['mpg'] - df['mpg'].min())/(df['mpg'].max() - df['mpg'].min())
```

```python
y, X = dmatrices('mpg ~ C(year)', data=df, return_type='dataframe')
logit = sm.Logit(y,X)
resultCat = logit.fit()
resultCat.summary2()
```
